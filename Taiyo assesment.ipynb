```python
import networkx as nx

# Create an empty graph
G = nx.Graph()
```

```python
# Iterate through rows in the DataFrame and add nodes and edges
for index, row in df.iterrows():
    # Extract relevant information from the row
    government = row['countryname']
    company = row['impagency']

    # Add nodes (entities) if they don't already exist
    if not G.has_node(government):
        G.add_node(government)

    if not G.has_node(company):
        G.add_node(company)

    # Add an edge (relationship) between the government and company
    G.add_edge(government, company)
```
